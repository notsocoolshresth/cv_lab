```python
import pandas as pd

# Load the CSV files
goodresplz = pd.read_csv('goodresplz.csv')
submission231 = pd.read_csv('submission231.csv')

# Compare values from index 1 to 300
for l in range(1, 300):
    if goodresplz.loc[l, 'Category'] == submission231.loc[l, 'Category']:
        goodresplz.loc[l, 'Category']= 'Rust'

# Save the modified DataFrame back to a new CSV file
goodresplz.to_csv('modified_goodresplz.csv', index=False)   
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load submission file
df = pd.read_csv("goodresplz.csv")

# Compute distribution
dist = df["Category"].value_counts()
total = len(df)

print("Total Samples:", total)
print("\nClass Distribution:\n")

for cls, count in dist.items():
    print(f"{cls} : {count} samples ({(count/total)*100:.2f}%)")

# Plot distribution
plt.figure()
dist.plot(kind="bar")
plt.title("Submission Prediction Distribution")
plt.ylabel("Count")
plt.xlabel("Category")
plt.show()
```
